Exploratory Data Analysis

Gender

```python
ax = sns.countplot(x = 'GENDER', data = df)

for bars in ax.containers:
    ax.bar_label(bars) 
```

From above graphs We can see the Most cancer patitent is males

SMOKING

Compare smoking to gender

```python
ax = sns.countplot(data = df, x = 'SMOKING', hue = 'GENDER')

sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

From above graphs We can see the 1 number chart is most smoking of  man and 2 number chart is most smoking of girls

YELLOW_FINGERS

compare Yellow_fingers to gender

```python
ax = df.groupby(['GENDER'],as_index = False)['YELLOW_FINGERS'].sum().sort_values(by = 'YELLOW_FINGERS' , ascending=False)
sns.set(rc={'figure.figsize':(4,4)})
plt.title("YELLOW_FINGERS")
sns.barplot( x = 'GENDER', y = 'YELLOW_FINGERS' , data = ax)
```

From above graphs We can see the Most case of  YELLOW_FINGERS are Males
